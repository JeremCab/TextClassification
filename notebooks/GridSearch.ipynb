```python
dataset_name = 'dbpedia_14'
```

```python
grid_search_d = {
    'dataset_name' : [dataset_name], # 'imdb', 'yelp_polarity', 'yelp_review_full', 
                      # 'trec', 'yahoo_answers_topics', 
                      # 'ag_news', 'dbpedia_14'],
    'tfidf_dim' : [1000, 2000, 3000, 4000],
    'pooling' : ['mean', 'mean_cls'], #, 'mean_std', 'cls', 'mean_cls', 'mean_std_cls'],
    'mode' : ['default'], #, 'bert_only', 'tfidf_only']
    'batch_size' : [64, 128, 256, 512],
    'sort' : [True, False]
}

keys, values = zip(*grid_search_d.items())

permutations_dicts = [dict(zip(keys, v)) for v in itertools.product(*values)]
```

```python
len(permutations_dicts)
```

**Note:**

Once the results are obtained, relauch the simulations in the ``bert_only`` and ``tfidf_only`` modes, in order to compare with the proposed combined method.

```python
progress_file = os.path.join(results_folder, f'{dataset_name}.progress')
progress_l = permutations_dicts.copy()
len(progress_l)
```

```python
for params_gs in tqdm(permutations_dicts):
    
    params_str = ''
    
    for k, v in params_gs.items():
        params_str += f" --{k} {v}"
    
    try:
        os.system("python3 experiment.py" + params_str) # new command!

        progress_l.pop(0)

        with open(progress_file, 'wb') as fh:
            pickle.dump(progress_l, fh)

        torch.cuda.empty_cache()
    except:
        pass
```
